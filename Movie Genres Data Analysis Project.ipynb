```python
import pandas as pd

movies = pd.read_csv(r'C:\Users\gugaa\Documents\Python\Pandas_Project 1_Movie Genres Data Analysis Project\imdb_movies.csv')
```

```python
movies.info()
```

```python
pd.set_option('display.max.rows',11000)
pd.set_option('display.max.columns',22)
```

```python
movies.head()
```

```python
# searching for duplicated values
movies[movies.duplicated()]
```

```python
#removing duplicated values
movies.drop_duplicates(inplace = True)
```

```python
movies.dropna(subset = ['genres'], inplace = True)
```

```python
movies['profit'] = movies['revenue'] - movies['budget']
```

```python
movies_genre = movies[['popularity','budget','revenue','original_title','runtime','genres','release_year','vote_count','vote_average','profit']]
```

```python
movies_genre.head()
```

```python
from pandas import Series, DataFrame
```

```python
split = movies_genre['genres'].str.split('|').apply(Series,1).stack()
split.index = split.index.droplevel(-1)
split.name = 'genres_split'
del movies_genre['genres']
movie_genre = movies_genre.join(split)
```

```python
movie_genre
```

### 1. Which genres are the most common (number of movies made)

```python
genres_count = pd.DataFrame(movie_genre.groupby('genres_split').original_title.nunique()).sort_values('original_title',ascending = True)
```

```python
genres_count['original_title'].plot.barh(title = 'Movies per Genre', color = 'darkblue')
```

### 2. Which genres have high avg. budget and revenue ?

```python
genres_avg = movie_genre.groupby('genres_split').mean(numeric_only = True)
pd.options.display.float_format = '{:2f}'.format
```

```python
genres_avg
```

```python
genres_avg.sort_values('budget', ascending = False, inplace = True)
```

```python
genres_avg[['budget','revenue']].plot.barh(title = 'Budget and Revenue by genre', color = ('darkblue', 'c'))
```

### 2.5 Which genres have high avg. budget and revenue ?

```python
genres_avg.sort_values('profit', ascending = False, inplace = True)
```

```python
genres_avg['profit'].plot.barh(title = 'Profit by genre', color = 'darkblue')
```

### 3.Which genres have high avg. popularity?

```python
genres_avg.sort_values('popularity', ascending = False, inplace = True)
```

```python
genres_avg['popularity'].plot.barh(title = 'popularity by genre', color = 'darkblue')
```

### 1. The best according to vote avg. return high profit and revenue.


```python
movies.drop_duplicates(inplace = True)
movies['profit'] = movies['revenue'] - movies['budget']
movies_genre = movies[['popularity','budget','revenue','original_title','runtime','genres','release_date','vote_count','vote_average','profit']]
```

```python
movies_counted = movies_genre[movies_genre['vote_count'] >= 50]
```

```python
movies_counted.corr(method = 'spearman',numeric_only = True)
```

```python
import seaborn as sns
sns.regplot(x = 'vote_average', y = 'profit', data = movies_counted, line_kws = {'color':'red'})
```

### 2. The best movies according to popularity return high profit and revenue


```python
movies_counted.corr(method = 'spearman',numeric_only = True)
```

```python
sns.regplot(x = 'popularity', y = 'revenue', data = movies_counted, line_kws = {'color':'red'})
```

### 3. Highly budgeted movies return high profit.


```python
import matplotlib.pyplot as plt
sns.regplot(x = 'budget', y = 'profit', data = movies_counted, line_kws = {'color':'red'})
plt.show()
```

### 4. Highly budgeted movies have a high popularity

```python
sns.regplot(x = 'budget', y = 'popularity', data = movies_counted, line_kws = {'color':'red'})
plt.show()
```

### 5. Highly profit per genre per year

```python
movie_genre.head()
```

```python
time_genre = pd.DataFrame(movie_genre.groupby(['release_year','genres_split'])['profit'].mean())
```

```python
final_genre = pd.pivot_table(time_genre,values = 'profit', index = ['genres_split'], columns = ['release_year'])
```

```python
sns.set(rc = {'figure.figsize' : (15,10)})
sns.heatmap(final_genre, cmap = 'coolwarm', linewidths =.5)
plt.title('Genres by Profit per year')
```
